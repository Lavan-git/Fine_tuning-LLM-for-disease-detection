## Data Preprocessing and Feature Engineering

### Subtask:
Clean and preprocess the 'DiseaseAndSymptoms' dataset by handling missing symptom values and combining multiple symptom columns into a single, structured 'Symptoms' column. Clean and preprocess the 'Disease precaution' dataset by handling missing values and ensuring consistent formatting. Finally, merge both datasets on the 'Disease' column.


**Reasoning**:
First, I will preprocess the 'DiseaseAndSymptoms' DataFrame by handling missing values in the symptom columns, stripping whitespace, converting to lowercase, concatenating all symptom columns into a single 'Symptoms' column, dropping the original symptom columns, and removing duplicate rows.



```python
df_symptoms_processed = df_symptoms.copy()

# 1. Fill missing values (NaN) in 'Symptom_' columns with empty strings
symptom_cols = [col for col in df_symptoms_processed.columns if 'Symptom_' in col]
for col in symptom_cols:
    df_symptoms_processed[col] = df_symptoms_processed[col].fillna('')

# 2. Clean the symptom data: strip leading/trailing whitespace and convert to lowercase
for col in symptom_cols:
    df_symptoms_processed[col] = df_symptoms_processed[col].astype(str).str.strip().str.lower()

# 3. Create a new column 'Symptoms' by concatenating all cleaned 'Symptom_' columns
#    Ensure no empty symptom strings are included and join with ', '
df_symptoms_processed['Symptoms'] = df_symptoms_processed[symptom_cols].apply(
    lambda row: ', '.join(row[row != '']), axis=1
)

# 4. Drop the original individual 'Symptom_1' through 'Symptom_17' columns
df_symptoms_processed = df_symptoms_processed[['Disease', 'Symptoms']]

# 5. Remove any duplicate rows based on 'Disease' and 'Symptoms'
df_symptoms_processed.drop_duplicates(subset=['Disease', 'Symptoms'], inplace=True)

print("Processed df_symptoms head:")
print(df_symptoms_processed.head())
print("\nProcessed df_symptoms info:")
df_symptoms_processed.info()
print("\nProcessed df_symptoms missing values:")
print(df_symptoms_processed.isnull().sum())
```

**Reasoning**:
Next, I will preprocess the 'df_precaution' DataFrame by handling missing values, stripping whitespace, converting to lowercase, and concatenating all precaution columns into a single 'Precautions' column. Finally, I will merge the processed 'df_symptoms_processed' and 'df_precaution_processed' DataFrames.



```python
df_precaution_processed = df_precaution.copy()

# 6. Fill missing values (NaN) in 'Precaution_3' and 'Precaution_4' columns with empty strings
precaution_cols = [col for col in df_precaution_processed.columns if 'Precaution_' in col]
for col in precaution_cols:
    df_precaution_processed[col] = df_precaution_processed[col].fillna('')

# 7. Clean the precaution data: strip leading/trailing whitespace and convert to lowercase
for col in precaution_cols:
    df_precaution_processed[col] = df_precaution_processed[col].astype(str).str.strip().str.lower()

# 8. Create a new column 'Precautions' by concatenating all cleaned 'Precaution_' columns
#    Ensure no empty precaution strings are included and join with ', '
df_precaution_processed['Precautions'] = df_precaution_processed[precaution_cols].apply(
    lambda row: ', '.join(row[row != '']), axis=1
)

# Drop the original individual 'Precaution_1' through 'Precaution_4' columns
df_precaution_processed = df_precaution_processed[['Disease', 'Precautions']]

print("Processed df_precaution head:")
print(df_precaution_processed.head())
print("\nProcessed df_precaution info:")
df_precaution_processed.info()
print("\nProcessed df_precaution missing values:")
print(df_precaution_processed.isnull().sum())

# 9. Perform an inner merge of the processed df_symptoms and df_precaution DataFrames
df_merged = pd.merge(df_symptoms_processed, df_precaution_processed, on='Disease', how='inner')

print("\nMerged DataFrame head:")
print(df_merged.head())
print("\nMerged DataFrame info:")
df_merged.info()
print("\nMerged DataFrame missing values:")
print(df_merged.isnull().sum())
```

## Generate JSONL Datasets

### Subtask:
Transform the merged and preprocessed data into the specified JSONL format. Each entry will include 'instruction', 'input' (symptoms), and 'output' (disease, explanation, and a medical safety disclaimer). Split the formatted data into `train.jsonl` and `test.jsonl` files.


**Reasoning**:
To prepare the data for LLM fine-tuning, I will first define the instruction and disclaimer strings, then iterate through the merged DataFrame to format each entry with 'instruction', 'input' (symptoms), and a JSON 'output' containing 'disease', 'explanation' (from precautions), and the medical safety 'disclaimer'. I will also import necessary libraries.



```python
import json
from sklearn.model_selection import train_test_split

# 1. Define the fixed instruction string
instruction_string = "You are a medical assistant chatbot. Based on the symptoms provided, predict the disease and suggest precautions."

# 2. Define a fixed medical safety disclaimer string
disclaimer_string = "Disclaimer: This information is for educational purposes only and should not be used as a substitute for professional medical advice. Always consult with a qualified healthcare provider for any health concerns."

# 3. Create an empty list to store the formatted JSONL entries
formatted_data = []

# 4. Iterate through each row of the df_merged DataFrame
for index, row in df_merged.iterrows():
    disease = row['Disease']
    symptoms = row['Symptoms']
    precautions = row['Precautions']

    # 5. Construct the 'input' string
    input_string = f"Symptoms: {symptoms}"

    # Construct the explanation string
    explanation = f"The predicted disease is {disease}. To prevent {disease}, you should {precautions}."

    # 6. Construct the 'output' dictionary
    output_dict = {
        "disease": disease,
        "explanation": explanation,
        "disclaimer": disclaimer_string
    }

    # 7. Convert this 'output' dictionary to a JSON string
    output_json_string = json.dumps(output_dict)

    # 8. Create a dictionary for the current entry and append it to the list
    entry = {
        "instruction": instruction_string,
        "input": input_string,
        "output": output_json_string
    }
    formatted_data.append(entry)

print(f"Generated {len(formatted_data)} formatted entries. Here's the first one:")
print(formatted_data[0])

# 9. Split the list of formatted entries into training and testing sets
# Using an 80/20 split as an example
train_data, test_data = train_test_split(formatted_data, test_size=0.2, random_state=42)

print(f"\nTrain set size: {len(train_data)}")
print(f"Test set size: {len(test_data)}")

# 10. Save the training set to train.jsonl
with open('train.jsonl', 'w') as f:
    for entry in train_data:
        f.write(json.dumps(entry) + '\n')

# 11. Save the testing set to test.jsonl
with open('test.jsonl', 'w') as f:
    for entry in test_data:
        f.write(json.dumps(entry) + '\n')

print("\n'train.jsonl' and 'test.jsonl' created successfully.")
```

## LoRA/QLoRA Configuration and Fine-Tuning

### Subtask:
Configure LoRA/QLoRA parameters for efficient fine-tuning. Train the loaded LLM on the `train.jsonl` dataset for at least 2 epochs, carefully managing batch sizes and gradient accumulation. Monitor training progress and save the fine-tuned adapter/model.


The next step is to prepare the model and data for fine-tuning. This involves configuring the tokenizer to handle padding, setting up the LoRA configuration for efficient adaptation of the Gemma-2B model, loading the `train.jsonl` dataset, and defining the training parameters using `TrainingArguments`. These steps will ensure that the model is ready for the fine-tuning process.

**Reasoning**:
Now I will configure the tokenizer, set `model.config.use_cache`, define the LoRA configuration, load the `train.jsonl` dataset, and set up the training arguments to prepare for fine-tuning the model.



```python
from peft import LoraConfig
from datasets import load_dataset
from transformers import TrainingArguments
from trl import SFTTrainer

# 1. Configure the tokenizer with a padding token
# If the tokenizer doesn't have a padding token, set it to the EOS token
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token

# Set model.config.use_cache = False for training efficiency
model.config.use_cache = False

# 2. Define the LoRA configuration
lora_config = LoraConfig(
    r=8,  # LoRA attention dimension
    lora_alpha=16,  # Alpha parameter for LoRA scaling
    target_modules=["q_proj", "v_proj", "k_proj", "o_proj"], # Target modules for LoRA
    lora_dropout=0.05,  # Dropout probability for LoRA layers
    bias="none",  # Bias type for LoRA layers
    task_type="CAUSAL_LM"  # Task type for causal language modeling
)

# 3. Load the train.jsonl dataset
dataset = load_dataset('json', data_files='train.jsonl', split='train')

# Define a formatting function to prepare the data for SFTTrainer
def formatting_function(example):
    # Combine instruction, input, and the desired output (JSON string) into a single text
    # This format teaches the model to generate the structured JSON output based on the prompt.
    formatted_text = f"<s>[INST] {example['instruction']}\n{example['input']} [/INST]\n{example['output']}</s>"
    return formatted_text

# 4. Define the training arguments
training_arguments = TrainingArguments(
    output_dir="./results",  # Directory to save the fine-tuned model and logs
    per_device_train_batch_size=4,  # Batch size per GPU/CPU for training
    gradient_accumulation_steps=2,  # Number of updates steps to accumulate before performing a backward/update pass
    warmup_steps=50,  # Number of steps for the learning rate warmup
    num_train_epochs=2,  # Total number of training epochs
    learning_rate=2e-4,  # Initial learning rate for AdamW optimizer
    fp16=True,  # Use mixed precision training (recommended for T4 GPU)
    logging_steps=10,  # Log every X updates steps
    optim="paged_adamw_8bit",  # Optimizer to use, paged_adamw_8bit is memory efficient
    save_strategy="epoch",  # Save the model after each epoch
    push_to_hub=False, # Do not push model to Hugging Face Hub
    report_to="none" # Disable reporting to any experiment tracking platform
)

# 5. Initialize the SFTTrainer
trainer = SFTTrainer(
    model=model,
    train_dataset=dataset,
    peft_config=lora_config,
    args=training_arguments,
    formatting_func=formatting_function # Use the formatting function instead of dataset_text_field
)

# 6. Start the fine-tuning process
print("Starting model fine-tuning...")
trainer.train()

print("Fine-tuning complete.")
print("Saving fine-tuned adapter...")
trainer.save_model("./fine_tuned_gemma_adapter")
print("Fine-tuned adapter saved to ./fine_tuned_gemma_adapter")
```

## Model Prediction and Evaluation Setup

### Subtask:
Load the `test.jsonl` dataset and use the fine-tuned model to make predictions. Extract the predicted disease labels from the model's output to prepare for evaluation.


**Reasoning**:
To prepare for model prediction, I will first load the `test.jsonl` dataset using the `datasets` library and then initialize a `pipeline` for text generation from the `transformers` library, specifying the fine-tuned model and tokenizer.



```python
from datasets import load_dataset
from transformers import pipeline
import json

# 1. Load the test.jsonl dataset
test_dataset = load_dataset('json', data_files='test.jsonl', split='train')

# 2. Initialize a pipeline for text generation
# Use the fine-tuned model and its tokenizer
generator = pipeline(
    'text-generation',
    model=model,
    tokenizer=tokenizer
)

# 3. Create a list to store the predicted disease labels
predicted_diseases = []

print(f"Loaded test dataset with {len(test_dataset)} entries.")
print("Text generation pipeline initialized.")
```

**Reasoning**:
Now that the text generation pipeline is initialized, I will iterate through the `test_dataset`, construct the prompt using the `instruction` and `input` fields, generate text using the `generator` pipeline, and then parse the model's output to extract the predicted disease, storing it in the `predicted_diseases` list.



```python
import re

# 4. Iterate through each entry in the loaded test dataset and make predictions
for i, entry in enumerate(test_dataset):
    instruction = entry['instruction']
    input_text = entry['input']

    # 5. Form the prompt for the model using the same format as during training
    prompt = f"<s>[INST] {instruction}\n{input_text} [/INST]\n"

    # 6. Use the text generation pipeline to get the model's prediction
    # Set max_new_tokens to allow the model to generate the full structured output
    # Use do_sample=True for more varied outputs, or False for deterministic outputs
    # Use temperature for controlling randomness
    # Stop generation when the '</s>' token is encountered
    outputs = generator(
        prompt,
        max_new_tokens=512,  # Adjust as needed to capture full output
        do_sample=True,
        temperature=0.7,
        top_k=50,
        top_p=0.95,
        eos_token_id=tokenizer.eos_token_id
    )

    # Extract the generated text
    generated_text = outputs[0]['generated_text']

    # The generated text will contain the prompt and the model's output.
    # We need to extract only the model's response part.
    # The expected output format is '[INST] ... [/INST]\n{JSON_output}</s>'
    # So, we look for the part after the last '[/INST]\n'
    try:
        # Find the start of the actual JSON output
        output_start_idx = generated_text.rfind('[/INST]\n')
        if output_start_idx != -1:
            # + len('[/INST]\n') to start from after the delimiter
            json_str_with_suffix = generated_text[output_start_idx + len('[/INST]\n'):]
            # Remove the '</s>' suffix if present
            json_str_clean = json_str_with_suffix.replace('</s>', '').strip()

            # Attempt to parse the cleaned JSON string
            # Sometimes the model might generate extra text around the JSON, so we need to be robust
            # Use regex to find the most likely JSON string within the response
            match = re.search(r'\{.*?\}', json_str_clean)
            if match:
                json_str = match.group(0)
                parsed_output = json.loads(json_str)
                predicted_disease = parsed_output.get('disease')
                predicted_diseases.append(predicted_disease)
            else:
                print(f"Warning: Could not find valid JSON in output for entry {i}. Output: {json_str_clean}")
                predicted_diseases.append(None) # Append None for unparsable outputs
        else:
            print(f"Warning: Could not find '[/INST]\n' in generated text for entry {i}. Full output: {generated_text}")
            predicted_diseases.append(None) # Append None if pattern is not found
    except json.JSONDecodeError as e:
        print(f"JSON decoding error for entry {i}: {e}. Generated text part: {json_str_clean if 'json_str_clean' in locals() else generated_text}")
        predicted_diseases.append(None) # Append None for JSON errors
    except Exception as e:
        print(f"An unexpected error occurred for entry {i}: {e}. Generated text part: {generated_text}")
        predicted_diseases.append(None)

print(f"Completed predictions for {len(predicted_diseases)} test entries.")
print(f"First 5 predicted diseases: {predicted_diseases[:5]}")
```

**Reasoning**:
Now that predictions have been made for the test dataset, the next step is to evaluate the model's performance by comparing these predictions against the true labels. This involves extracting the true disease labels from the `test.jsonl` dataset and aligning them with the generated predictions. This preparation is essential for generating evaluation metrics, such as a confusion matrix.



```python
import json

# 1. Extract the true disease labels from the test_dataset
true_diseases = []
for entry in test_dataset:
    # The true output is a JSON string, so we need to parse it
    true_output_json = json.loads(entry['output'])
    true_diseases.append(true_output_json['disease'])

# 2. Filter out None values from predicted_diseases to ensure consistent lengths
# This step is crucial if some predictions failed to parse
filtered_true_diseases = []
filtered_predicted_diseases = []

for i, pred in enumerate(predicted_diseases):
    if pred is not None and true_diseases[i] is not None:
        filtered_predicted_diseases.append(pred)
        filtered_true_diseases.append(true_diseases[i])
    else:
        print(f"Skipping entry {i} due to missing prediction or true label.")

print(f"Extracted {len(true_diseases)} true disease labels.")
print(f"Number of valid predictions/true labels for evaluation: {len(filtered_predicted_diseases)}")
print(f"First 5 true diseases: {filtered_true_diseases[:5]}")
```

## Generate and Visualize Confusion Matrix

### Subtask:
Compare the actual disease labels from `test.jsonl` with the predicted disease labels to create a confusion matrix. Generate an image of this confusion matrix, ensuring all necessary labels and a legend are included, as required for submission.


**Reasoning**:
To generate and visualize the confusion matrix, I will import the necessary libraries, extract unique disease labels, compute the confusion matrix, and then use seaborn and matplotlib to plot and save it according to the specified instructions.



```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Get a list of all unique disease labels present in both true and predicted lists
all_labels = sorted(list(set(filtered_true_diseases + filtered_predicted_diseases)))

# 2. Calculate the confusion matrix
cm = confusion_matrix(filtered_true_diseases, filtered_predicted_diseases, labels=all_labels)

# 3. Create a figure and an axes object for the plot
plt.figure(figsize=(18, 15)) # Increased figsize to accommodate more labels

# 4. Use seaborn.heatmap to visualize the confusion matrix
sns.heatmap(cm,
            annot=True, # Show the numbers in each cell
            fmt='d', # Format as integers
            cmap='Blues', # Color map
            xticklabels=all_labels, # Labels for x-axis
            yticklabels=all_labels) # Labels for y-axis

# 5. Add a title to the plot
plt.title('Confusion Matrix of Disease Predictions', fontsize=16)

# 6. Label the x-axis and y-axis
plt.xlabel('Predicted Label', fontsize=14)
plt.ylabel('True Label', fontsize=14)

# 7. Rotate the x-axis labels for better readability
plt.xticks(rotation=90, ha='right', fontsize=10)
plt.yticks(rotation=0, fontsize=10) # Keep y-labels horizontal

# Ensure the layout is tight to prevent labels from overlapping
plt.tight_layout()
plt.savefig('confusion_matrix1.png')
# 8. Display the plot
plt.show()

# 9. Save the confusion matrix plot as an image file
#plt.savefig('confusion_matrix1.png')

print("Confusion matrix generated, displayed, and saved as 'confusion_matrix1.png'.")
```

## Execute and Format Demo Query

### Subtask:
Run the specified demo query (e.g., 'Symptoms: Fever, headache, body pain') through the fine-tuned LLM. Ensure the model's response strictly adheres to the required structured format, including the explanation and the medical safety disclaimer.


**Reasoning**:
To demonstrate the fine-tuned model's capability, I will define a demo input string, construct the prompt, use the `generator` pipeline to get the model's prediction, and then extract and print the structured JSON output.



```python
import json, re

# 1. Define demo inputs
demo_symptoms_list = [
    'Symptoms: Fever, headache, body pain',          # Required test case
    'Symptoms: Fever, cough, sore throat',           # Case 2
    'Symptoms: itching, skin rash, nodal skin eruptions'  # Case 3
]

# List to store all results
demo_results = []

# 2. Loop through test cases
for demo_symptoms in demo_symptoms_list:
    print("\n" + "="*60)
    print(f"Generating prediction for: {demo_symptoms}")
    print("="*60)

    # Construct prompt
    demo_prompt = f"<s>[INST] {instruction_string}\n{demo_symptoms} [/INST]\n"

    # Generate response
    demo_outputs = generator(
        demo_prompt,
        max_new_tokens=512,
        do_sample=True,
        temperature=0.7,
        top_k=50,
        top_p=0.95,
        eos_token_id=tokenizer.eos_token_id
    )

    generated = demo_outputs[0]['generated_text']

    parsed_output = None
    raw_section = ""

    # Parse JSON
    try:
        idx = generated.rfind('[/INST]\n')
        if idx != -1:
            after_inst = generated[idx + len('[/INST]\n'):].replace('</s>', '').strip()
            raw_section = after_inst
            match = re.search(r'\{.*?\}', after_inst, re.DOTALL)

            if match:
                json_str = match.group(0)
                parsed_output = json.loads(json_str)
                print("\nParsed JSON Output:")
                print(json.dumps(parsed_output, indent=2))
            else:
                print("\nWarning: JSON not found. Showing raw output:")
                print(after_inst)
        else:
            print("\nWarning: '[/INST]' not found. Full output:")
            print(generated)
            raw_section = generated

    except Exception as e:
        print("\nUnexpected error while parsing:", e)
        raw_section = generated

    # Store output in list
    demo_results.append({
        "query": demo_symptoms,
        "parsed_output": parsed_output,
        "raw_output": raw_section
    })

# 3. Save all results into a JSON file
with open("demo_outputs.json", "w") as f:
    json.dump(demo_results, f, indent=2)

print("\nSaved all demo outputs to demo_outputs.json 🎉")
```
